```python
import numpy as np
import scipy
from sympy import Matrix
from sympy.physics.quantum import TensorProduct
from toqito.matrix_props import is_hermitian
```

```python
#su(4)
list = []
s0 = np.array([[1,0],[0,1]])
s1 = np.array([[0,1],[1,0]])
s2 = np.array([[0,-1j],[1j,0]])
s3 = np.array([[1,0],[0,-1]])
list.append(s0)
list.append(s1)
list.append(s2)
list.append(s3)
#print('list of matrices :',list)
matrix_list = []
for i in list:
    for j in list:
        x = TensorProduct(i,j)
        matrix_list.append(x)
        # print(x, "\n")

su4_basis = []

for i in matrix_list[1:]:
    su4_basis.append(i)

# find the non-zero positions of all elements in the basis
for i in range(len(su4_basis)):
    print("Basis element ", i+1, np.nonzero(su4_basis[i]))
```

```python
# check whether basis is hermitian

j = 1
for i in su4_basis:
    print("Basis Element", j)
    print("Is matrix Hermitian? ",is_hermitian(i), "\n")
    j = j+1
```

```python
# check whether the trace condition is satisfied
# Tr$(\lambda_i \lambda_j) = 2\delta_{ik}$

for i in range(len(su4_basis)):
    for j in range(len(su4_basis)):
        mat_prod = np.matmul(su4_basis[i], su4_basis[j])
        if i == j:
            print("Element ", i+1, "with itself has trace", np.trace(mat_prod))

for i in range(len(su4_basis)):
    print("\n")
    for j in range(len(su4_basis)):
        mat_prod = np.matmul(su4_basis[i], su4_basis[j])
        if i != j:
            print("Element ", i+1, "with Element ",j+1,"has trace", np.trace(mat_prod))
```

```python
# find the third element in the commutator
# find the matrices that commute with each other and non-zero positions of those that do
for i in range(len(su4_basis)):
    print("\n")
    for j in range(len(su4_basis)):
        if i != j:
            comm_ij = commutator(su4_basis[i], su4_basis[j])
            if np.array_equal(comm_ij, np.zeros((4,4))):
                print("[ Lambda ",i+1,", Lambda ", j+1," ] = 0")
            for k in range(len(su4_basis)):
                basis_times_i = 1j*2*su4_basis[k]
                diff1 = comm_ij - basis_times_i
                diff2 = comm_ij + basis_times_i
                if np.array_equal(diff1, np.zeros((4,4))):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = 2i Lambda", k+1)
                elif np.array_equal(diff2, np.zeros((4,4))):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = -2i Lambda", k+1)
                elif np.array_equal(np.nonzero(comm_ij), np.nonzero(su4_basis[k])):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = Some linear combination with the 3 diagonal matrices or Lambda", k+1)
```

```python
# find the commuting elements i.e zero structure constants
for i in range(len(su4_basis)):
    print("\n")
    for j in range(len(su4_basis)):
        print("\n")
        for k in range(len(su4_basis)):
            if i != j != k:
                comm_ij = commutator(su4_basis[i], su4_basis[j])
                comm_ij_times_k = np.matmul(comm_ij, su4_basis[k])
                tr_ijk = np.trace(comm_ij_times_k)
                if tr_ijk == 0:
                    print("Elements", i+1, j+1, k+1, "commute")
```

```python
# define the commutator
def commutator(a_matrix, b_matrix):
    diff = np.matmul(a_matrix, b_matrix) - np.matmul(b_matrix, a_matrix)
    return diff

# find the third element in the commutator
# find the matrices that commute with each other and non-zero positions of those that do
for i in range(len(so6_basis)):
    print("\n")
    for j in range(len(so6_basis)):
        if i != j:
            comm_ij = commutator(so6_basis[i], so6_basis[j])
            for k in range(len(so6_basis)):
                basis_times_i = 1j*so6_basis[k]
                diff1 = comm_ij - basis_times_i
                diff2 = comm_ij + basis_times_i
                if np.array_equal(diff1, np.zeros((6,6))):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = i Lambda", k+1)
                elif np.array_equal(diff2, np.zeros((6,6))):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = -i Lambda", k+1)
                elif np.array_equal(np.nonzero(comm_ij), np.nonzero(so6_basis[k])):
                    print("[ Lambda ",i+1,", Lambda ", j+1," ] = Some linear combination with Lambda", k+1)
```
